```python
from sklearn.pipeline import make_pipeline
```

```python
# Используем Pipeline для масштабирования и поиска по сетке для Logreg
```

```python
# Logistic Regression pipeline and hyperparameter grid
pipe_logreg = make_pipeline(MinMaxScaler(), LogisticRegression(max_iter=1000))
param_grid_logreg = {
    'logisticregression__C': [0.01, 0.1, 1, 10, 100],
    'logisticregression__penalty': ['l1', 'l2'],
    'logisticregression__solver': ['liblinear', 'saga']
}
# Grid search for Logistic Regression
grid_logreg = GridSearchCV(pipe_logreg, param_grid_logreg, cv=5, scoring='accuracy')
grid_logreg.fit(X_train, y_train)
```

```python
# Используем Pipeline для масштабирования и поиска по сетке для SVM
```

```python
# SVM pipeline and hyperparameter grid
pipe_svm = make_pipeline(StandardScaler(), SVC())
param_grid_svm = [
    {'svc__kernel': ['linear'], 'svc__C': [0.1, 1, 10, 100]},
    {'svc__kernel': ['rbf'], 'svc__C': [0.1, 1, 10, 100], 'svc__gamma': ['scale', 'auto']}
]

# Grid search for SVM
grid_svm = GridSearchCV(pipe_svm, param_grid_svm, cv=5, scoring='accuracy')
grid_svm.fit(X_train, y_train)

# Predictions and reports
y_pred_logreg = grid_logreg.predict(X_test)
y_pred_svm = grid_svm.predict(X_test)
```

```python
# Смотрим результаты
```

```python
report_logreg = classification_report(y_test, y_pred_logreg, output_dict=True)
report_svm = classification_report(y_test, y_pred_svm, output_dict=True)
```

```python
# Prepare results summary
results = {
    'Logistic Regression': {
        'best_params': grid_logreg.best_params_,
        'best_score': grid_logreg.best_score_,
        'test_classification_report': report_logreg
    },
    'SVM': {
        'best_params': grid_svm.best_params_,
        'best_score': grid_svm.best_score_,
        'test_classification_report': report_svm
    }
}
```
